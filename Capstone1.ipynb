```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Consumer_Complaints.csv', on_bad_lines='skip')

df.info()
df.head()
```

```python
df.describe(include='all')
```

```python
df['Date received'] = pd.to_datetime(df['Date received'])

complaints_by_date = df.groupby('Date received').size()

plt.figure(figsize=(10, 5))
complaints_by_date.plot()
plt.title('Number of Complaints Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
print(df.columns)
```

```python
complaints_by_product = df['Product'].value_counts()

plt.figure(figsize=(10, 5))
complaints_by_product.plot(kind='bar')
plt.title('Number of Complaints by Product')
plt.xlabel('Product')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
complaints_by_company = df['Company'].value_counts().head(10)

plt.figure(figsize=(10, 5))
complaints_by_company.plot(kind='bar')
plt.title('Number of Complaints by Company')
plt.xlabel('Company')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])
df['Resolution time (days)'] = (df['Date sent to company'] - df['Date received']).dt.days

top_companies = df['Company'].value_counts().head(10).index
resolution_times_by_company = df[df['Company'].isin(top_companies)]

plt.figure(figsize=(12, 6))
resolution_times_by_company.boxplot(column='Resolution time (days)', by='Company', rot=90)
plt.title('Resolution Time by Company')
plt.xlabel('Company')
plt.ylabel('Resolution Time (days)')
plt.suptitle('')
plt.show()
```
